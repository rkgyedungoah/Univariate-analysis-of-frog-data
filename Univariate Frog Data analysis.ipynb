```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import first frog data

frog_data1 = pd.read_excel('all_frog_data.xlsx')
frog_data1
```

```python
# import second frog data

frog_data2 = pd.read_excel('All_MLOut_new.xlsx')
frog_data2
```

```python
# column names of second frog data 

frog_data2.columns
```

```python
# merging the two dataframes

frog_data = frog_data1.merge(frog_data2[['gbifID', 'occurrenceStatus', 'key', 'ppt_mean', 'soil_mean', 'tmax_mean', 'tmin_mean']], on = 'gbifID', how = 'inner')
frog_data.head()
```

```python
# deleting the unnamed column

frog_data = frog_data.drop('Unnamed: 0', axis = 1)
frog_data
```

```python
frog_data.head()
```

Two dataframes have been merged to achieve a single one which contains the same number of rows. Univariate analysis of each column is next.


### ppt mean

```python
# histogram representation of ppt_mean

plt.hist(frog_data['ppt_mean'])
plt.show()
```

```python
# description of ppt_mean

frog_data['ppt_mean'].describe()
```

It can be seen that the 75th percentile is 77.9. This implies that most (75%) of the frogs live in a habitat with low precipitation (0.77 - 77.9). Only a few (25%) of the frogs live in a habitat with high precipitation values (77.9 - 368.75).

### Soil Mean

```python
# histogram representation of soil_mean

plt.hist(frog_data['soil_mean'])
plt.show()
```

```python
# description of soil_mean

frog_data['soil_mean'].describe()
```

The bulk (75%) of the frogs live in a habitat with low soil mean values (0.00 - 69.07). A few (25%) live in habitats with high soil mean values between 69.07 and 403.72. This means most of the frogs live in habitats with low soil moisture.

### tmax mean

```python
# histogram representation of tmax_mean

plt.hist(frog_data['tmax_mean'])
plt.show()
```

```python
# description of tmax_mean

frog_data['tmax_mean'].describe()
```

The 75th percentile is 24.78 indicating that most of the frogs live in habitats with temperatures within room temperature or below it. A few (25%) live in habitats with relatively high temperatures between 24.2 and 37.4.

```python
# histogram representation of tmin_mean

plt.hist(frog_data['tmin_mean'])
plt.show()
```

```python
# description of tmin_mean

frog_data['tmin_mean'].describe()
```

75% of the frogs live in habitats with minimum temperatures between 0.8 and 12.7. 25% live in habitats with minimum temperatures between 12.7 and 25.3. This indicates that the frogs live in cold habitats.
